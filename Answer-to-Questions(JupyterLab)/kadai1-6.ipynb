```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng= np.random.default_rng();
```

```python
N=10000
mu=-3
sigma=2
xx= rng.normal(loc=mu,scale=sigma,size=N)
```

```python
plt.hist(xx,density=True)

x= np.linspace(mu-5*sigma,mu+5*sigma,1000)
y= 1/((2*math.pi)**0.5*sigma)*np.exp(-(x-mu)**2/(2*sigma**2))

plt.plot(x,y)
```
