## 課題1-10 回答例

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng= np.random.default_rng();
```

```python
lb,ub= -1,1
N= 10000
x= np.empty(N)
x[0]= rng.uniform(low=lb,high=ub)

a= 0.9
for i in range(1,N):
    x[i]= a*x[i-1] + rng.uniform(low=lb,high=ub)
```

```python
plt.figure(figsize=(12,4))
plt.plot(range(N),x)
```

```python
# 系列データの不偏分散値
np.var(x,ddof=1)
```

```python
eps= 1e-4 #展開を打ち切る係数の閾値
n= int(np.log(eps)/np.log(a)) #項数
n
```

```python
#一般化中心極限定理による、系列データの母分散の推定値
CLT_var= 0
for i in range(n):
    CLT_var+= (ub*a**i-lb*a**i)**2/12
```

```python
CLT_var
```

```python
plt.hist(x,bins=50, density=True)

mu= 0
sigma = CLT_var**0.5
x_clt= np.linspace(mu-5*sigma,mu+5*sigma,1000)
y_clt= 1/((2*math.pi)**0.5*sigma)*np.exp(-(x_clt-mu)**2/(2*sigma**2))

plt.plot(x_clt,y_clt)
```
